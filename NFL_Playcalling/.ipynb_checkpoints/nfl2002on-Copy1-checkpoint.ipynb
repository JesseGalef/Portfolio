```python
def detect_rush(desc):
    desc = desc.lower()
    direction = re.search('\. ?\w+ (\w+)', desc)
    if direction != None: 
        if direction.group(1) in ['left', 'right','up','to', 'scrambles']: return 1
    return 0
```

```python
def yards_net(desc):
    desc = desc.lower()
    for verb in ['fumble','intercept','field goal','punt','kick']:
        if verb in desc: return 0
    num= re.search('(-?[0-9]+?) yard', desc)
    if num == None:
        return 0
    else:
        num = int(num.group(1))
        if ('loss of' in desc)|(('penalty' in desc)&('decline' not in desc)&('between downs' not in desc)):
            return -num
        else:
            return num
    return 0
```

```python
def cv_predict_proba(estimator, X, y, params={}, n_folds=3, verbose=1):
    kfold = KFold(n=len(X),n_folds=n_folds, shuffle=True, random_state=42)
    probs = pd.Series(0, index=y.index)
    if params != {}:
        cv_kfold = KFold(n=len(X),n_folds=3, shuffle=True, random_state=42)
        grid = GridSearchCV(estimator, param_grid=params, scoring='roc_auc', cv=cv_kfold, verbose=verbose)
        if verbose>3: print 'gridsearching estimator hyperparameters'
        grid.fit(X,y)
        estimator = grid.best_estimator_

    for i, (train, test) in enumerate(kfold):
        if verbose >3: print 'getting probabilities for section', str(i)
        estimator.fit(X.iloc[train], y.iloc[train])
        probs.iloc[test] = estimator.predict_proba(X.iloc[test])[:,1]
    if verbose > 0:
        print "roc_auc score of", roc_auc_score(list(y), list(probs))
    return probs
```

```python
def cv_get_multiclass_weight(estimator, X, d, n_folds=3, verbose=1):
    kfold = KFold(n=len(X),n_folds=n_folds, shuffle=True, random_state=42)
    probs = pd.DataFrame(0, index=d.index, columns=sorted(d.unique()))

    for i, (train, test) in enumerate(kfold):
        if verbose >3: print 'getting probabilities for section', str(i)
        probs.iloc[test] = estimator.fit(X.iloc[train], d.iloc[train]).predict_proba(X.iloc[test])
    weights = pd.Series(0,index=d.index)
    for col in probs.columns:
        weights.loc[d==col] = (1/probs[d==col][col])

    return weights, probs
```

```python
def get_weighted_df(df, treatment='incomplete', target='nextrushAttempt', propensities = '', cols_to_use='', params='', estimator=GradientBoostingClassifier(), verbose=0, n_folds=20):
    if type(propensities) == pd.Series:
        probs = propensities
        weights = (df[treatment]/propensities)+(1.-df[treatment])/(1.-propensities)
    else:
        if cols_to_use == '':
            cols_to_use = ['qtr','min','ydline','lead','hurrying','runOutClock', 'togo',
                   'team_rush_avg', 'def_team_rush_avg',
                  'team_rush_gains', 'team_pass_gains', 'team_rush_def', 'team_pass_def']
        X = df[cols_to_use]
        d = df[treatment]
        if params != '':
            kfold = KFold(n=len(X),n_folds=3, shuffle=True, random_state=42)
            grid = GridSearchCV(estimator, param_grid=params, scoring='roc_auc', cv=kfold, verbose=verbose)
            if verbose>1: print 'gridsearching estimator hyperparameters'
            grid.fit(X,d)
            estimator = grid.best_estimator_
    #     if verbose >1:
    #         print 'getting cross-validated propensities'
    #     propensities = (list(cv_predict_proba(estimator, X, d, n_folds=n_folds, verbose=verbose)), index=d.index)
        weights, probs = cv_get_multiclass_weight(estimator, X, d, n_folds=n_folds, verbose=verbose)

#     if verbose >0: print "roc_auc score:", str(roc_auc_score(list(d), list(propensities)))
#     weights = (df[treatment]/propensities)+(1.-df[treatment])/(1.-propensities)
#     weights = weights * weights.count() / weights.sum()
    
    return pd.crosstab(df[target], df[treatment], weights,
            aggfunc=sum)
```

```python
# http://nbviewer.jupyter.org/github/kellieotto/StatMoments/blob/master/PSM.ipynb
def Match(groups, propensity, caliper = 0.05):
    ''' 
    Inputs:
    groups = Treatment assignments.  Must be 2 groups
    propensity = Propensity scores for each observation. Propensity and groups should be in the same order (matching indices)
    caliper = Maximum difference in matched propensity scores. For now, this is a caliper on the raw
            propensity; Austin reccommends using a caliper on the logit propensity.
    
    Output:
    A series containing the individuals in the control group matched to the treatment group.
    Note that with caliper matching, not every treated individual may have a match.
    '''

    # Check inputs
    if any(propensity <0) or any(propensity >1):
        raise ValueError('Propensity scores must be between 0 and 1')
    elif not(0<caliper<1):
        raise ValueError('Caliper must be between 0 and 1')
    elif len(groups)!= len(propensity):
        raise ValueError('groups and propensity scores must be same dimension')
    elif len(groups.unique()) != 2:
        raise ValueError('wrong number of groups')
        
        
    # Code groups as 0 and 1
    groups = groups == groups.unique()[0]
    N = len(groups)
    N1 = groups.sum(); N2 = N-N1
    g1, g2 = propensity[groups == 1], propensity[groups == 0]
    # Check if treatment groups got flipped - treatment (coded 1) should be the smaller
    if N1 > N2:
       N1, N2, g1, g2 = N2, N1, g2, g1 
        
        
    # Randomly permute the smaller group to get order for matching
    morder = np.random.permutation(N1)
    matches = pd.Series(np.empty(N1))
    matches[:] = np.NAN
    
    for m in morder:
        dist = abs(g1.iloc[m] - g2) # Create a Series of the distances of each point in group2 to the m-th point in group1
        if dist.min() <= caliper:
            matches[m] = dist.argmin()
            g2 = g2.drop(matches[m]) # JG - Comment this out to use replacement
    return (matches)
```

```python
def match_df(match_df, treatment, propensity='', target, caliper=.01, verbose=1):

    # Takes a dataframe with inputted treatment, target, and propensity column names
    # Uses the Match function to create (and return) a dataframe 'matched' with each
    # data point in the treatment group paired with a point in the non-treatment group
    # of similar propensity to be treated
    
    # The DataFrame 'observed' totals the counts of 'target' within each of the treatment
    # and non-treatment groups within 'matched'

    
    #     if treatment == None: treatment = match_df['incomplete']
#     if propensity == None: propensity = match_df['propensity']
#     if target == None: target = match_df['nextrushAttempt']
        

    matching = Match(treatment, propensity, caliper=caliper)
    index_matching = matching.dropna().astype(int)

    flipped = treatment.value_counts()[1]/float(len(treatment))>.5

    group1 = target[treatment==flipped].loc[(index_matching.values).astype(int)]
    group2 = target[treatment!=flipped].iloc[(index_matching.index).astype(int)]

    if flipped:
        temp_group = group1
        group1 = group2
        group2 = temp_group

    observed = pd.DataFrame()
    observed['After Not incomplete'] = group1.value_counts(sort=False) # These are hard coded, ah well
    observed['After Incomplete'] = group2.value_counts(sort=False)
    observed.index= ["Not Rush Next","Rush Next"]
    matched = match_df[treatment==(~flipped)].iloc[(index_matching.index).astype(int)]
    matched = matched.append(match_df[treatment==(flipped)].loc[(index_matching.values).astype(int)])
#     weighted_matched_df = weight_df(match_df, treatment, propensity, target)
    if verbose >1:
        print 
        print str(1-observed.sum().sum() / float(min(treatment.value_counts())*2)), "of the treatment data points could not be matched"
    
    return observed, matched
```

```python
def weight_df(df, treatment='incomplete', propensity='propensity', target='nextrushAttempt'):
    weights = (df[treatment]/df[propensity])+(1.-df[treatment])/(1.-df[propensity])
    weights = weights * weights.count() / weights.sum()
    return pd.crosstab(df[target], df[treatment], weights,
            aggfunc=sum)
    
def adjusted_df(df, treatment='Incomplete', weight = 'weight', target='adjusted'):
    weighted_df = pd.DataFrame(index=['Pass Next','Rush Next'], columns=['NotAfterIncomplete','AfterIncomplete'])

    weighted_df.ix[1,0] = df[df[treatment]==0][target].sum()
    weighted_df.ix[0,0] = df[df[treatment]==0][weight].sum()-df[df[treatment]==0][target].sum()
    weighted_df.ix[1,1] = df[df[treatment]==1][target].sum()
    weighted_df.ix[0,1] = df[df[treatment]==1][weight].sum()-df[df[treatment]==1][target].sum()
    return weighted_df
```
